# Baseline model

```python
df_train
```

```python
X, y = df_train.iloc[:, :-1], df_train['Strength']

def train_eval(X, y, model, model_param):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

    cv = KFold(n_splits=2)
    grid_model = GridSearchCV(estimator=model,
                              param_grid=model_param,
                              scoring='neg_mean_squared_error'
                             )

    grid_model.fit(X_train, y_train)
    y_val_pred = grid_model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))

    print('RMSE :', rmse)
```

```python
param_lgbm = {'n_estimators': [50, 100, 200],
              'max_depth':[-1, 5, 7, 9, 11],
#               'num_leaves':[40, 60, 80],
              'min_child_samples':[5,10,20],
              'reg_alpha':[0, 0.01],
              'random_state':[99]}

param_xgb = {'n_estimators': [50, 100, 200],
             'max_depth':[3, 5, 7, 9],
#              'num_leaves':[30, 60],
             'gamma': [0.5, 1, 2],
#              'learning_rate ': ['eta', 0.01, 0.1],
             'random_state':[99]}
```

```python
train_eval(X, y, LGBMRegressor(), param_lgbm)
```

```python
train_eval(X, y, XGBRegressor(), param_xgb)
```
